```python
import yfinance as yf
```

### Baixando cotações


```python
#baixando dados do IBOV 

ibov = yf.download('^BVSP')
```

```python
ibov
```

```python
# Baixando cotações de ITUB4, entre 2017 e 2023
itub4 = yf.download('ITUB4.SA', start='2017-01-01', end='2022-12-31')
```

```python
itub4
```

```python
# Baixando cotações dos últimos 12 meses de PETR4 
petr4 = yf.download('PETR4.SA', period='1y')
```

```python
petr4
```

```python
# Baixando dados intraday, minuto a minuto de VALE3, dos últimos 5 pregões
#m = minutos/ y = ano / wk = semana / mo = mês

# Valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
# Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
vale3 = yf.download('VALE3.SA', interval='1m', period='5d')
```

```python
vale3
```

```python
# Baixando várias empresa de uma vez. 
# Tickers separados por espaço

yf.download("WEGE3.SA BBDC4.SA")
```

```python
# Baixando várias empresa de uma vez.
# Passando uma lista de tickers

yf.download(['WEGE3.SA','BBDC4.SA', 'PETR4.SA'])
```

```python
# Alterando o método de agrupamento
# Group by 'ticker' or 'column' (default)

yf.download(['WEGE3.SA','BBDC4.SA', 'PETR4.SA'], group_by='ticker')
```
